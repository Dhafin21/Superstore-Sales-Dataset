5. Category Product Analysis

```python
kategoriproduk_sales = (
    df.groupby(["Category"]) ["Sales"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

kategoriproduk_order = (
    df.groupby(["Category"]) ["Order ID"]
    .nunique()
    .sort_values(ascending=False)
    .reset_index()
)

kategoriproduk_all = pd.merge(
    kategoriproduk_sales,
    kategoriproduk_order,
    on=["Category"]
)

kategoriproduk_all.head()
```

```python
kategoriproduk_display = kategoriproduk_all.copy()

kategoriproduk_display["Sales"] = kategoriproduk_display["Sales"].map('${:,.2f}'.format)

kategoriproduk_display
```

-> Produk yang terjual terdapat 3 kategori yaitu Technology, Furniture, dan Office Supplies

-> Dari 3 kategori sales terbesar ada pada kategori Technology tetapi memiliki penjualan yang sangat sedikit

-> sedangkan untuk kategori office supplies memiliki penjualan terkecil tetapi memiliki penjualan yang terbanyak bahkan 2 kali lipat lebihh dari kategori technology

Bisa disimpulkan bahwa kategori Technology merupakan kategori dengan total penjualan terbesar dengan nilai $827.455,87, walaupun hanya mencatat sebanyak 1.519 transaksi. 
Sedangkan, Office Supplies memiliki jumlah transaksi tertinggi yaitu 3.676 order, tetapi total penjualannya merupakan yang paling rendah dibanding kategori lainnya. Hal ini menunjukkan bahwa pelanggan cenderung membeli produk Office Supplies lebih sering, sementara transaksi pada kategori Technology memiliki nilai pembelian yang lebih besar pada setiap order.

6. Sub Category Product Analysis

```python
subkategoriproduk_sales = (
    df.groupby(["Sub-Category"]) ["Sales"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

subkategoriproduk_order = (
    df.groupby(["Sub-Category"]) ["Order ID"]
    .nunique()
    .sort_values(ascending=False)
    .reset_index()
)

subkategoriproduk_all = pd.merge(
    subkategoriproduk_sales,
    subkategoriproduk_order,
    on=("Sub-Category")
)

subkategoriproduk_all
```

7. PRODUCT ANALYSIS

```python
produk_sales = (
    df.groupby(["Product Name", "Sub-Category", "Category"]) ["Sales"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

produk_order = (
    df.groupby(["Product Name","Sub-Category", "Category"]) ["Order ID"]
    .nunique()
    .sort_values(ascending=False)
    .reset_index()
)

produksales_all = pd.merge(
    produk_sales,
    produk_order,
    on=["Product Name","Sub-Category", "Category"]
)

produkorder_all = pd.merge(
    produk_order,
    produk_sales,
    on=["Product Name", "Sub-Category", "Category"]
)
```

```python
produkorder_all.head(10)
```

```python
produk_display = produkorder_all.copy()

produk_display["Sales"] = produk_display["Sales"].map('${:,.2f}'.format)

produk_display
```
