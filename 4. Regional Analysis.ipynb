5. Sales by Region

```python
region_sales = (
    df.groupby(["Region"]) ["Sales"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

region_order = (
    df.groupby(["Region"]) ["Order ID"]
    .nunique()
    .sort_values(ascending=False)
    .reset_index()
)

region_customer = (
    df.groupby(["Region"]) ["Customer ID"]
    .nunique()
    .sort_values(ascending=False)
    .reset_index()
)

```

```python
region_all = (
    region_sales
    .merge(region_order, on="Region")
    .merge(region_customer, on="Region")
)

# Menghitung kontribusi (%)
region_all["Contribution (%)"] = (
    region_all["Sales"] / region_all["Sales"].sum() * 100
).round(2)

# Menghitung rata rata pembelian
region_all["Average"] = (
    region_all["Sales"] / region_all["Order ID"]
)
region_all
```

```python
region_sales_display = region_all.copy()

region_sales_display["Sales"] = region_sales_display["Sales"].map("${:,.2f}".format)

region_sales_display["Average"] = region_sales_display["Average"].map("${:,.2f}".format)

region_sales_display
```

-> Region dengan jumlah penjualan terbanyak, Order terbanyak, dan Kontribusi terbesar ada di region West dengan nilai Sales $710,219.68, order sebanyak 1587, dan kontribusi sebesar 31.40% tidak hanya itu saja region West memiliki jumlah customer terbanyak diantara region yang lainnya.

-> Region dengan jumlah penjualan terkecil, Order terkecil, dan Kontribusi terkecil ada di region South dengan nilai Sales $389,151.46, order sebanyak 810, dan kontribusi sebesar 17.21%

6. Sales by State

```python
state_sales = (
    df.groupby(["Region","State"]) ["Sales"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

state_order = (
    df.groupby(["Region","State"]) ["Order ID"]
    .nunique()
    .sort_values(ascending=False)
    .reset_index()
)

state_all = pd.merge(
    state_sales,
    state_order,
    on=["Region", "State"]
)

# Menghitung kontribusi (%)
state_all["Contribution (%)"] = (
    state_all["Sales"] / state_all["Sales"].sum() * 100
).round(2)

state_all.head(10)
```

```python
state_display = state_all.copy()

state_display["Sales"] = state_display["Sales"].map("${:,.2f}".format)

state_display.head(10)
```

```python
state_display.tail(10)
```

-> State yang terbanyak dari jumlah order dan penjualan terdapat di State California dengan jumlah order 1.002, serta dengan nilai penjualan sebesar $446,306.46, kontribusi 19.73%

-> State dengan nominasi terkecil ada di state Nort Dakota jumlah order hanya 2, besar penjualan $919.91, dan kontribusi yang diberikan hanya  3.12%

7. Sales by City

```python
city_sales = (
    df.groupby(["Region","State","City"]) ["Sales"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

city_order = (
    df.groupby(["Region","State","City"]) ["Order ID"]
    .nunique()
    .sort_values(ascending=False)
    .reset_index()
)

city_all = pd.merge(
    city_sales,
    city_order,
    on=["Region", "State", "City"]
)

# Menghitung kontribusi (%)
city_all["Contribution (%)"] = (
    city_all["Sales"] / city_all["Sales"].sum() * 100
).round(2)

city_all.head(10)
```

```python
city_display = city_all.copy()

city_display["Sales"] = city_display["Sales"].map("${:,.2f}".format)

city_display.head(10)
```

```python
city_display.tail(10)
```

-> New York merupakan Kota yang menyumbang penjualan terbanyak dengan nilai $252,462.55, memiliki order 439 paling besar diantara lainnya dan memiliki kontribusi sebesar 11.16%

-> Sebaliknya Kota bernama Abilene menyumbang penjualan sebesar $1.39 dengan order sebanyak 1 kali saja, banyak yg memiliki order hanya sekali tetapi nilai sales diatas dari kota Abilene
